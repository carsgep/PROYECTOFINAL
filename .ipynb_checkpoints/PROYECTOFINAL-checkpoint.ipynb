```python
plt.figure(figsize=(8,4))
by_date = frame >> group_by(X.FECHA) >> summarize(Cases = X.FECHA.count())
sns.lineplot(x = 'FECHA', y = 'Cases', data = by_date, palette='viridis')

"""
Existe un corte en la serie de tiempo; es decir hay dos períodos claramente distintos. 
Esto sugiere usar solo un periodo
"""
```

```python
frame['YEAR'] = frame['FECHA'].dt.year
frame['DAY'] = frame['FECHA'].dt.weekday_name
frame['MONTH'] = frame['FECHA'].dt.month
frame['WEEK'] = frame['FECHA'].dt.week
```

```python
plt.figure(figsize=(8,4))
by_date = frame >> group_by(X.YEAR) >> summarize(Cases = X.FECHA.count())
sns.barplot(x = 'YEAR', y = 'Cases', data = by_date, palette='viridis')

"""
Por qué a partir de 2014 aumenta el número de casos reportados?
La serie siempre es creciente.
"""
```

```python
plt.figure(figsize=(8,4))
by_date = frame >> group_by(X.MONTH) >> summarize(Cases = X.FECHA.count())
sns.barplot(x = 'MONTH', y = 'Cases', data = by_date, palette='viridis')
"""
El último trimestre se registran menos casos, por qué?
Mayo es el mes más violento (Día de la Madre)
"""
```

```python
plt.figure(figsize=(8,4))
order = ['Monday', 'Tuesday','Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
by_date = frame >> group_by(X.DAY) >> summarize(Cases = X.FECHA.count())
sns.barplot(x = 'DAY', y = 'Cases', data = by_date, palette='viridis', order=order)

"""
El domingo es el día que se presenta el mayor número de casos. De hecho, los casos aumentan desde el viernes.
"""
```

```python
plt.figure(figsize=(12,6))
by_date = frame >> group_by(X.WEEK) >> summarize(Cases = X.FECHA.count())
sns.barplot(x = 'WEEK', y = 'Cases', data = by_date, palette='viridis')

"""
Sinusoidal hasta la semana 41
"""
```

```python
plt.figure(figsize=(20,6))
by_date = frame >> group_by(X.DEPARTAMENTO) >> summarize(Cases = X.FECHA.count()) >> ungroup() >>arrange(X.Cases, ascending= False)
plt.xticks(rotation = 90)
sns.barplot(x = 'DEPARTAMENTO', y = 'Cases', data = by_date, palette='viridis')
```
